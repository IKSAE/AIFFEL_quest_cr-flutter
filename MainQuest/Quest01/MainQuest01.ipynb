# 메인 퀘스트 1번: 은행계좌 만들기

```python
import random

class Account:
    # 클래스 변수
    bank_name = "SC은행"
    account_count = 0

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.account_number = self.generate_account_number()
        self.deposit_count = 0
        self.deposit_history_list = []
        self.withdraw_history_list = []
        Account.account_count += 1

    # 계좌 번호 생성
    def generate_account_number(self):
        return f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}"

    # 클래스 메서드: 생성된 계좌 개수 출력
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # 입금 메서드
    def deposit(self, amount):
        if amount >= 1:
            old_balance = self.balance
            self.balance += amount
            self.deposit_count += 1
            self.deposit_history_list.append((self.deposit_count, amount, old_balance, self.balance))
            print(f"{amount:,}원이 입금되었습니다. 현재 잔액: {self.balance:,}원")
            # 이자 지급 (입금 5회마다 1% 이자 지급)
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += int(interest)
                self.deposit_history_list.append((self.deposit_count, interest, old_balance, self.balance))
                print(f"이자가 {interest:,.0f}원 지급되었습니다. 현재 잔액: {self.balance:,}원")
        else:
            print("입금은 최소 1원 이상이어야 합니다.")

    # 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
        elif amount >= 1:
            old_balance = self.balance
            self.balance -= amount
            self.withdraw_history_list.append((len(self.withdraw_history_list) + 1, amount, old_balance, self.balance))
            print(f"{amount:,}원이 출금되었습니다. 현재 잔액: {self.balance:,}원")
        else:
            print("출금은 최소 1원 이상이어야 합니다.")

    # 계좌 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {Account.bank_name}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,}원")

    # 입금 내역 출력
    def deposit_history(self):
        print(f"{self.owner}님의 입금 내역:")
        for count, amount, old_balance, new_balance in self.deposit_history_list:
            action = "입금"
            if amount != old_balance:  # If this is an interest payment
                action = "이자지급"
            print(f"{count}회: {action} 금액: {amount:,}원 잔액: {new_balance:,}원")

    # 출금 내역 출력
    def withdraw_history(self):
        print(f"{self.owner}님의 출금 내역:")
        for count, amount, old_balance, new_balance in self.withdraw_history_list:
            print(f"{count}회: 출금 금액: {amount:,}원 잔액: {new_balance:,}원")

# 테스트 코드

# 계좌 생성
first_account = Account("차정은", 1000)
second_account = Account("박광석", 30000000)
third_account = Account("조웅제", 50000000)

# 클래스 변수 확인 (생성된 계좌의 개수 출력)
print(f"현재까지 생성된 계좌 수: {Account.get_account_num()}개")

# 잔고 100만원 이상인 고객 정보 출력
print("\n잔고 100만원 이상인 고객 정보:")
accounts = [first_account, second_account, third_account]
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()

# 입금 및 출금 테스트
my_account = Account("차정은", 1000)
for i in range(10):
    my_account.deposit(1000)
my_account.withdraw(300000)
my_account.withdraw(3000)

# 입금 내역 및 출금 내역 출력
print('-' * 50)
my_account.deposit_history()
print('-' * 50)
my_account.withdraw_history()
```

# 메인 퀘스트 2번: 간단한 자동사냥 RPG 만들기

```python
import random

class Character:
    def __init__(self, name, level, health, attack, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        net_damage = max(0, damage - self.defense)
        self.health -= net_damage

    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f"{self.name}이 {target.name}에게 {damage} 만큼 공격했다...!")
        target.take_damage(damage)
```

```python
class Player(Character):
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0

    def gain_experience(self, amount):
        self.experience += amount
        self.level_up()

    def level_up(self):
        while self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            print(f"레벨업! 현재 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}")

class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack, defense)
```

```python
def battle(player, monster):
    print(f"{monster.name}과의 전투를 시작합니다.")
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
            print(f"{monster.name}의 체력: {monster.health}")
            print(f"{player.name}의 체력: {player.health}")

    if player.is_alive():
        print(f"전투 승리! {monster.name}을 이겼다!")
        player.gain_experience(monster.level * 20)
    else:
        print("전투 패배.. 게임오버")
```

```python
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    player_name = input("플레이어의 이름을 입력하세요: ")
    player = Player(player_name)

    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)
        battle(player, monster)
        if not player.is_alive():
            print("게임오버")
            break

if __name__ == "__main__":
    main()
```
